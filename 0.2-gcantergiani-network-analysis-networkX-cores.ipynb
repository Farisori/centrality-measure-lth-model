```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH = "../data/raw/"

FILE_NETWORK_SOCIAL_EXAMPLE = "higgs-social_network_example.edgelist"
```

```python
data = pd.read_csv(PATH + FILE_NETWORK_SOCIAL_EXAMPLE, sep=' ', names = ['source', 'target'])
data.head()
```

```python
G = nx.from_pandas_dataframe(data, 'source', 'target')
```

```python
#k_core(G, k=None, core_number=None)[source]
#Return the k-core of G.
#A k-core is a maximal subgraph that contains nodes of degree k or more.

k_core_graph = nx.k_core(G)
nx.draw(k_core_graph, with_labels = True, node_size=700)
plt.show()
```

```python
df2 = nx.to_pandas_dataframe(k_core_graph)
df2
```

```python
#Return the k-shell of G.

#The k-shell is the subgraph of nodes in the k-core but not in the (k+1)-core.

#Parameters:	
#G (NetworkX graph) – A graph or directed graph.
#k (int, optional) – The order of the shell. If not specified return the main shell.
#core_number (dictionary, optional) – Precomputed core numbers for the graph G.

k_shell_graph = nx.k_shell(G)
nx.draw(k_shell_graph, with_labels = True, node_size=700)
plt.show()
```

```python
df3 = nx.to_pandas_dataframe(k_shell_graph)
df3
```
