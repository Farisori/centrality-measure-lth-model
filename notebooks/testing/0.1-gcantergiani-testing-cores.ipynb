```python
%matplotlib inline 

import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
G=nx.dodecahedral_graph()
nx.draw(G)
```

```python
print('nx.core_number(G) : {}'.format(nx.core_number(G)))
print()

k=max(nx.core_number(G).values())
print('max(nx.core_number(G).values()): {}'.format(max(nx.core_number(G).values())))

k_core = nx.k_core(G)
nx.draw(k_core)
```

```python
G=nx.barbell_graph(10,10)

print('nx.core_number(G) : {}'.format(nx.core_number(G)))
print()

k=max(nx.core_number(G).values())
print('max(nx.core_number(G).values()): {}'.format(max(nx.core_number(G).values())))

nx.draw(G)
```

```python
k_core = nx.k_core(G)

print('nx.core_number(G) : {}'.format(nx.core_number(G)))
print()

k=max(nx.core_number(G).values())
print('max(nx.core_number(G).values()): {}'.format(max(nx.core_number(G).values())))

nx.draw(k_core)
```

```python
for node in k_core:
    print('Node {} | degree {}'.format(node, G.degree(node)))
```

```python
G=nx.gnm_random_graph(100,500, directed = True)

print('nx.core_number(G) : {}'.format(nx.core_number(G)))
print()

k=max(nx.core_number(G).values())
print('max(nx.core_number(G).values()): {}'.format(max(nx.core_number(G).values())))

nx.draw(G, edge_color='b', node_size = 30)
```

```python
k_core = nx.k_core(G)

print('nx.core_number(G) : {}'.format(nx.core_number(G)))
print()

k=max(nx.core_number(G).values())
print('max(nx.core_number(G).values()): {}'.format(max(nx.core_number(G).values())))

nx.draw(k_core)
```

```python
for node in k_core:
    print('Node {}  \t degree {}'.format(node, G.degree(node)))
    print('\t\t in_degree {}'.format( G.in_degree(node)))
    print('\t\t out_degree {}'.format( G.out_degree(node)))
    print()
```
